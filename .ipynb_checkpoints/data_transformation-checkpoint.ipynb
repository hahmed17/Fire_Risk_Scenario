# Landscape Data Compilation

Given landscape data in `TIF` and `GEOJSON` format, use FARSITE's ``lcpmake`` executable to compile ``LCP`` file format for FARSITE compatibility. 

Eight landscape raster bands are required to run FARSITE: slope, aspect, fire behavior fuel model, tree canopy cover, canopy height, canopy base height, and canopy bulk density.

### Confirm Data and Executables Filepaths 

```python
TIFF_DIR = BASE_DIR / 'Forest'
LCPMAKE_EXEC_PATH = FARSITE_DIR / 'lcpmake'

# Check that input data exists
print("Input data locations:")
if not os.path.exists(TIFF_DIR):
    print("WARNING: Missing landscape TIFF files.")
else:
    print("Landscape layers (.tif) directory: ", TIFF_DIR)
    
if not os.path.exists(LCPMAKE_EXEC_PATH):
    print("WARNING: Missing LCPMAKE executable to compile .LCP file.")
else:
    print("lcpmake executable path: ", LCPMAKE_EXEC_PATH)
```

### Create output location for transformed landscape data

```python
LCP_OUTPUT_DIR = BASE_DIR / 'Forest_LCP_Outputs'
LCP_OUTPUT_PATH = LCP_OUTPUT_DIR / 'landscape.lcp'

# Create output DIRECTORY if not already exist
Path(LCP_OUTPUT_DIR).mkdir(parents=True, exist_ok=True)  # Create output directory if diesn't exist

print("\nOutput data locations:")
print("ASCII files (.asc) and landscape file (.lcp) output directory: ", LCP_OUTPUT_PATH)
print("Landscape filepath ", LCP_OUTPUT_PATH)
```

### Raster band filepaths

```python
# Existing raster files
ELEV_PATH = TIFF_DIR / 'Forest_elevation.tif'
FBFM_PATH = TIFF_DIR / 'Forest_fbfm13.tif'
TREELIST_PATH = TIFF_DIR / 'Forest_Treelist.json'

# Missing raster files
SLPP_PATH = LCP_OUTPUT_DIR / "slope.tif"
ASP_PATH = LCP_OUTPUT_DIR / "aspect.tif"
HT_PATH = LCP_OUTPUT_DIR / "canopy_height.tif"
CBH_PATH = LCP_OUTPUT_DIR / "canopy_base_height.tif"
CBD_PATH = LCP_OUTPUT_DIR / "canopy_bulk_density.tif"
DIA_PATH = LCP_OUTPUT_DIR / "maximum_canopy_diameter.tif"
CC_PATH = LCP_OUTPUT_DIR / "canopy_cover.tif"


print("Given landscape layers:")
print("Elevation: ", ELEV_PATH)
print("Fire Behavior Fuel Model: ", FBFM_PATH)
print('Tree data (Canopy Height, Canopy Base Height, Canopy Bulk Density): ', TREELIST_PATH)
print("\nMissing layers: Slope, Aspect, Canopy Height, Canopy Base Height, Canopu Bulk Density, Canopy Cover")
```

## Derive Missing Raster Bands from Landscape Data

```python
gdal.DEMProcessing(str(SLPP_PATH), str(ELEV_PATH), "slope", slopeFormat="percent")
gdal.DEMProcessing(str(ASP_PATH),  str(ELEV_PATH), "aspect")

ref_ds = gdal.Open(str(ELEV_PATH))
gt     = ref_ds.GetGeoTransform()
nx     = ref_ds.RasterXSize
ny     = ref_ds.RasterYSize
proj   = ref_ds.GetProjection()
ref_ds = None

x_min = gt[0];  y_max = gt[3]
x_max = x_min + nx * gt[1]
y_min = y_max + ny * gt[5]

lat_rad      = math.radians(IGNITION_LAT)
px_w_m       = abs(gt[1]) * 111320 * math.cos(lat_rad)
px_h_m       = abs(gt[5]) * 111320
cell_area_m2 = px_w_m * px_h_m

# align fuel to elevation grid using CLI — handles missing CRS correctly
FBFM_PATH_RESIZED = LCP_OUTPUT_DIR / "fbfm13.tif"
result = subprocess.run([
    "gdalwarp",
    "-s_srs", "WGS84",
    "-t_srs", "WGS84", 
    "-te",  str(x_min), str(y_min), str(x_max), str(y_max),
    "-ts",  str(nx), str(ny),
    "-r",   "near",
    "-srcnodata", "9999",
    "-dstnodata", "-9999",
    "-of",  "GTiff",
    "-overwrite",
    str(FBFM_PATH),
    str(FBFM_PATH_RESIZED),
], capture_output=True, text=True)

if result.returncode != 0:
    raise RuntimeError(f"gdalwarp failed:\n{result.stderr}")

# verify
ds   = gdal.Open(str(FBFM_PATH_RESIZED))
data = ds.GetRasterBand(1).ReadAsArray()
print(f"FBFM resized: {ds.RasterXSize} x {ds.RasterYSize}  unique: {np.unique(data)}")
ds   = None

# rasterize canopy layers
for out_path, attribute in [
    (HT_PATH,  "HT"),
    (CBH_PATH, "CBH"),
    (CBD_PATH, "CBD"),
]:
    gdal.Rasterize(
        str(out_path),
        str(TREELIST_PATH),
        format="GTiff",
        attribute=attribute,
        outputBounds=[x_min, y_min, x_max, y_max],
        width=nx,
        height=ny,
        noData=0,
        outputType=gdal.GDT_Float32,
    )
    print(f"✓ {out_path.name}")

gdal.Rasterize(
    str(CC_PATH),
    str(TREELIST_PATH),
    format="GTiff",
    attribute="DIA",
    outputBounds=[x_min, y_min, x_max, y_max],
    width=nx,
    height=ny,
    noData=0,
    outputType=gdal.GDT_Float32,
)

ds   = gdal.Open(str(CC_PATH), gdal.GA_Update)
band = ds.GetRasterBand(1)
data = band.ReadAsArray().astype(np.float32)

valid       = data != 0
data[valid] = np.clip(
    (np.pi * (data[valid] / 2) ** 2) / cell_area_m2 * 100.0,
    0.0, 100.0
)

band.WriteArray(data)
band.SetNoDataValue(0)
band.FlushCache()
ds = None
print(f"✓ {CC_PATH.name}")
```

## Reproject `.tif` Files to EPSG 5070 (FARSITE compatibility)

```python
EPSG5070_WKT = """PROJCS["NAD83 / Conus Albers",
        GEOGCS["NAD83",
            DATUM["North_American_Datum_1983",
                SPHEROID["GRS 1980",6378137,298.257222101,
                    AUTHORITY["EPSG","7019"]],
                AUTHORITY["EPSG","6269"]],
            PRIMEM["Greenwich",0,
                AUTHORITY["EPSG","8901"]],
            UNIT["degree",0.0174532925199433,
                AUTHORITY["EPSG","9122"]],
            AUTHORITY["EPSG","4269"]],
        PROJECTION["Albers_Conic_Equal_Area"],
        PARAMETER["latitude_of_center",23],
        PARAMETER["longitude_of_center",-96],
        PARAMETER["standard_parallel_1",29.5],
        PARAMETER["standard_parallel_2",45.5],
        PARAMETER["false_easting",0],
        PARAMETER["false_northing",0],
        UNIT["metre",1,
            AUTHORITY["EPSG","9001"]],
        AXIS["Easting",EAST],
        AXIS["Northing",NORTH],
        AUTHORITY["EPSG","5070"]]"""

def reproject_tif(src_path, dst_path, resample_alg=gdal.GRA_Bilinear, nodata=0):
    gdal.Warp(
        str(dst_path),
        str(src_path),
        format="GTiff",
        srcSRS="WGS84",
        dstSRS=EPSG5070_WKT,
        resampleAlg=resample_alg,
        dstNodata=nodata,
    )
    print(f"✓ {Path(dst_path).name}")

# reproject each layer with appropriate nodata
for src, dst, nodata in [
    (ELEV_PATH,         ELEV_PATH_PROJ,  -9999),
    (SLPP_PATH,         SLPP_PATH_PROJ,  -9999),
    (ASP_PATH,          ASP_PATH_PROJ,   -9999),
    (FBFM_PATH_RESIZED, FBFM_PATH_PROJ,  -9999),  # -9999 not 0, since 0 is non-burnable
    (HT_PATH,           HT_PATH_PROJ,    0),
    (CBH_PATH,          CBH_PATH_PROJ,   0),
    (CBD_PATH,          CBD_PATH_PROJ,   0),
    (CC_PATH,           CC_PATH_PROJ,    0),
]:
    reproject_tif(src, dst, nodata=nodata)

# Projected .tif filepaths
ELEV_PATH_PROJ = LCP_OUTPUT_DIR / "elevation_5070.tif"
FBFM_PATH_PROJ = FBFM_PATH_RESIZED.with_stem(FBFM_PATH_RESIZED.stem  + "_5070")
SLPP_PATH_PROJ = SLPP_PATH.with_stem(SLPP_PATH.stem  + "_5070")
ASP_PATH_PROJ = ASP_PATH.with_stem(ASP_PATH.stem  + "_5070")
HT_PATH_PROJ = HT_PATH.with_stem(HT_PATH.stem  + "_5070")
CBH_PATH_PROJ = CBH_PATH.with_stem(CBH_PATH.stem  + "_5070")
CBD_PATH_PROJ = CBD_PATH.with_stem(CBD_PATH.stem  + "_5070")
CC_PATH_PROJ = CC_PATH.with_stem(CC_PATH.stem  + "_5070")

for src, dst in [
    (ELEV_PATH,  ELEV_PATH_PROJ, -9999),
    (SLPP_PATH,  SLPP_PATH_PROJ, -9999),
    (ASP_PATH,   ASP_PATH_PROJ, -9999),
    (FBFM_PATH_RESIZED,  FBFM_PATH_PROJ, -9999),
    (HT_PATH,    HT_PATH_PROJ),
    (CBH_PATH,   CBH_PATH_PROJ),
    (CBD_PATH,   CBD_PATH_PROJ),
    (CC_PATH,    CC_PATH_PROJ),
]:
    reproject_tif(src, dst, nodata=nodata)
```

## Convert `.tif` Bands to `.asc`

```python
# ASCII filepaths
ELEV_ASCII = LCP_OUTPUT_DIR / "elevation.asc"
SLPP_ASCII = SLPP_PATH.with_suffix(".asc")
ASP_ASCII  = ASP_PATH.with_suffix(".asc")
FBFM_ASCII = FBFM_PATH_RESIZED.with_suffix(".asc")
HT_ASCII   = HT_PATH.with_suffix(".asc")
CBH_ASCII  = CBH_PATH.with_suffix(".asc")
CBD_ASCII  = CBD_PATH.with_suffix(".asc")
CC_ASCII   = CC_PATH.with_suffix(".asc")
```

```python
tif_to_asc = [
    (ELEV_PATH_PROJ, ELEV_ASCII),
    (SLPP_PATH_PROJ, SLPP_ASCII),
    (ASP_PATH_PROJ,  ASP_ASCII),
    (FBFM_PATH_PROJ, FBFM_ASCII),
    (HT_PATH_PROJ,   HT_ASCII),
    (CBH_PATH_PROJ,  CBH_ASCII),
    (CBD_PATH_PROJ,  CBD_ASCII),
    (CC_PATH_PROJ,   CC_ASCII),
]

for tif_path, asc_path in tif_to_asc:
    gdal.Translate(
        str(asc_path),
        str(tif_path),
        format="AAIGrid",
    )
    print(f"✓ {asc_path.name}")
```

## Compile `.lcp` File Using `lcpmake` Executable

```python
lcp_filepath = LCP_OUTPUT_DIR / 'landscape'

cmd = [
    str(LCPMAKE_EXEC_PATH),
    "-latitude",  str(int(IGNITION_LAT)),
    "-landscape", str(lcp_filepath),
    "-elevation", str(ELEV_ASCII),
    "-slope",     str(SLPP_ASCII),
    "-aspect",    str(ASP_ASCII),
    "-fuel",      str(FBFM_ASCII),
    "-cover",     str(CC_ASCII),
    "-height",    str(HT_ASCII),
    "-base",      str(CBH_ASCII),
    "-density",   str(CBD_ASCII),
    "-fbfm40",
]

print("\nRunning lcpmake command:")
print(" ".join(cmd))

result = subprocess.run(cmd, capture_output=True, text=True)

print("stdout:", result.stdout)
print("stderr:", result.stderr)

if result.returncode != 0:
    raise RuntimeError(
        f"lcpmake failed with return code {result.returncode}\n"
        f"stdout: {result.stdout}\nstderr: {result.stderr}"
    )

lcp_filepath = lcp_filepath.with_suffix('.lcp')
print(f"\n✓ Landscape file created: {lcp_filepath}")
print(f"  Size: {lcp_filepath.stat().st_size / 1024:.1f} KB")
```

```python
# Just in case: add projection header to LCP file
def find_wkt_offset(lcp_path):
    with open(lcp_path, 'rb') as f:
        data = f.read()
    
    # find the large null region after the last .asc path
    last_asc = data.rfind(b'.asc')
    if last_asc == -1:
        raise RuntimeError("Could not find .asc marker in LCP")
    
    # scan forward from last .asc to find where nulls begin
    i = last_asc
    while i < len(data) and data[i] != 0:
        i += 1
    null_start = i
    
    # count the null run length
    j = i
    while j < len(data) and data[j] == 0:
        j += 1
    null_length = j - i
    
    print(f"Last .asc at offset: {last_asc}")
    print(f"Null region starts:  {null_start}")
    print(f"Null region length:  {null_length}")
    print(f"Bytes before nulls:  {repr(data[null_start-8:null_start])}")
    print(f"Bytes after nulls:   {repr(data[j:j+8])}")
    
    return null_start, null_length

offset, length = find_wkt_offset(lcp_filepath)

EPSG5070_WKT = 'PROJCS["NAD83 / Conus Albers",GEOGCS["NAD83",DATUM["North_American_Datum_1983",SPHEROID["GRS 1980",6378137,298.257222101]],PRIMEM["Greenwich",0],UNIT["degree",0.0174532925199433]],PROJECTION["Albers_Conic_Equal_Area"],PARAMETER["latitude_of_center",23],PARAMETER["longitude_of_center",-96],PARAMETER["standard_parallel_1",29.5],PARAMETER["standard_parallel_2",45.5],PARAMETER["false_easting",0],PARAMETER["false_northing",0],UNIT["metre",1]]'

wkt_bytes = EPSG5070_WKT.encode("utf-8")
if len(wkt_bytes) > length:
    raise ValueError(f"WKT too long: {len(wkt_bytes)} bytes (max {length})")

with open(lcp_filepath, 'r+b') as f:
    f.seek(offset)
    f.write(wkt_bytes.ljust(length, b'\x00'))

print(f"✓ Projection injected at offset {offset}")

# verify
with open(lcp_filepath, 'rb') as f:
    f.seek(offset)
    print(f"Verified: {f.read(len(wkt_bytes)).decode()}")
```
