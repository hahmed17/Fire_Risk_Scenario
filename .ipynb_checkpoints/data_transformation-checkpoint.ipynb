### Define Function to Convert `.tif` to `.asc`

```python
def convert_tif_to_ascii(
    input_path: str | Path,
    output_path: str | Path | None = None,
    band: int = 1,
    verbose: bool = False,
) -> Path:
    """
    Convert a single-band raster TIFF to an ASCII grid (.asc) file using GDAL.

    Args:
        input_path:  Path to the source .tif file.
        output_path: Path for the output .asc file. If None, uses the same
                     directory and stem as the input (e.g. slope.tif -> slope.asc).
        band:        Band index to export (1-based, default 1).
        verbose:     Print progress messages.

    Returns:
        Path to the written .asc file.

    Raises:
        FileNotFoundError: If the input file does not exist.
        RuntimeError:      If GDAL fails to open or translate the file.
    """
    input_path = Path(input_path)
    if not input_path.exists():
        raise FileNotFoundError(f"Input file not found: {input_path}")

    if output_path is None:
        output_path = input_path.with_suffix(".asc")
    output_path = Path(output_path)

    ds = gdal.Open(str(input_path))
    if ds is None:
        raise RuntimeError(f"GDAL could not open: {input_path}")

    n_bands = ds.RasterCount
    if band < 1 or band > n_bands:
        raise ValueError(f"Band {band} out of range — file has {n_bands} band(s).")
    ds = None  # close before Translate

    result = gdal.Translate(
        str(output_path),
        str(input_path),
        format="AAIGrid",
        bandList=[band],
    )

    if result is None:
        raise RuntimeError(f"GDAL Translate failed for: {input_path}")
    result = None  # flush and close

    if verbose:
        print(f"  ✓ {output_path.name}")

    return output_path
```

### Define Function to Confirm `.tif` and `.asc` Alignment

```python
def plot_tif_vs_asc(tif_path, asc_path, band=1, title=""):
    """
    Plot a TIFF band and its ASCII raster side-by-side to verify they match.
    """
    # --- read TIFF ---
    tif_ds = gdal.Open(str(tif_path))
    tif_data = tif_ds.GetRasterBand(band).ReadAsArray().astype(float)
    nodata = tif_ds.GetRasterBand(band).GetNoDataValue()
    if nodata is not None:
        tif_data[tif_data == nodata] = np.nan
    tif_ds = None

    # --- read ASC ---
    asc_ds = gdal.Open(str(asc_path))
    asc_data = asc_ds.GetRasterBand(1).ReadAsArray().astype(float)
    nodata = asc_ds.GetRasterBand(1).GetNoDataValue()
    if nodata is not None:
        asc_data[asc_data == nodata] = np.nan
    asc_ds = None

    # --- shared color scale ---
    vmin = np.nanmin([tif_data.min(), asc_data.min()])
    vmax = np.nanmax([tif_data.max(), asc_data.max()])

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(title or f"{Path(tif_path).stem} — TIF vs ASC", fontsize=13)

    kw = dict(cmap="terrain", vmin=vmin, vmax=vmax)

    im0 = axes[0].imshow(tif_data, **kw)
    axes[0].set_title("TIFF")
    plt.colorbar(im0, ax=axes[0], shrink=0.8)

    im1 = axes[1].imshow(asc_data, **kw)
    axes[1].set_title("ASC")
    plt.colorbar(im1, ax=axes[1], shrink=0.8)

    # difference plot — should be all zeros if they match
    diff = tif_data - asc_data
    im2 = axes[2].imshow(diff, cmap="RdBu", vmin=-1, vmax=1)
    axes[2].set_title(f"Difference (max abs: {np.nanmax(np.abs(diff)):.4f})")
    plt.colorbar(im2, ax=axes[2], shrink=0.8)

    plt.tight_layout()
    plt.show()

    # --- print a quick numeric summary ---
    print(f"TIF  — min: {np.nanmin(tif_data):.4f}  max: {np.nanmax(tif_data):.4f}")
    print(f"ASC  — min: {np.nanmin(asc_data):.4f}  max: {np.nanmax(asc_data):.4f}")
    print(f"Diff — max abs: {np.nanmax(np.abs(diff)):.6f}  mean abs: {np.nanmean(np.abs(diff)):.6f}")
```

```python
plot_tif_vs_asc(
    tif_path=str(LCP_DIR / 'Forest_elevation.tif'),
    asc_path=Path(asc),
    title="Elevation",
)
```
